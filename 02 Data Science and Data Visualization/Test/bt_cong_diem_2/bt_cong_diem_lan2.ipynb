```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb

weather = pd.read_csv('weather.csv')
weather.head()
```

```python
# Câu 1.1
box_data = weather[['TMIN', 'TAVG', 'TMAX']]
box_data.plot.box()
```

```python
# Câu 1.2
weather['TDIFF'] = weather['TMAX'] - weather['TMIN']
weather.head()
```

```python
# Câu 1.3
plt.hist(weather['TDIFF'],bins = 30)
plt.xlim(0)
plt.show()
plt.savefig('tdiff.png')
```

```python
# Câu 1.4
cols =['WT01', 'WT02', 'WT03', 'WT04', 'WT05', 'WT06', 'WT07', 'WT08', 'WT09', 'WT13', 'WT14', 'WT16', 'WT17', 'WT18', 'WT21', 'WT22']
bad_weather = pd.DataFrame(weather, columns = cols)
bad_weather
```

```python
# Câu 1.5
weather['bad_weather'] = bad_weather.sum(axis = 1)
weather
```

```python
# Câu 1.6
weather['bad_weather'] = weather['bad_weather'].fillna(0)
weather = weather.astype({'bad_weather': 'int'})
weather
```

```python
# Câu 1.7
plt.hist(weather['bad_weather'], bins = 10, color = 'green')
plt.title('Thời tiết')
plt.show()
```

```python
#Câu 2.1
movies = pd.read_csv('movie.csv')
sb.set_palette("Set2")
sb.set_style('whitegrid')
```

```python
#Câu 2.2
sb.jointplot(x = 'Rotten Tomatoes Ratings %', y='Audience Ratings %', data = movies)
plt.show()
```

```python
#Câu 2.3
plt.title('Audience Ratings')
sb.distplot(movies['Audience Ratings %'], bins = 15)
plt.show()
```

```python
#Câu 2.4
sb.lmplot(x = 'Rotten Tomatoes Ratings %', y='Audience Ratings %', hue = 'Genre', data = movies)
plt.show()
```

```python
#Câu 2.5
sb.set_palette('PRGn')
figure, axes = plt.subplots(nrows=1, ncols=2, figsize = (15,6))
sb.boxplot(x ='Genre', y = 'Rotten Tomatoes Ratings %', data = movies , ax=axes[0],)
sb.violinplot(x ='Genre', y = 'Rotten Tomatoes Ratings %', data = movies , ax=axes[1])
plt.show()
```
