```python
# Categorical & Categorical
# FullBath, BedroomAbvGr, TotRmsAbvGrd
# contigency table: Ho: FullBath and BedroomAbvGr independent

table_FB = pd.crosstab(df['FullBath'], df['BedroomAbvGr'])
table_FB
```

```python
table_FB.plot(kind='bar', stacked=True)
plt.show()
```

```python
# chi_squared test with similar proportions
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
# Chi square test
stat, p, dof, expected = chi2_contingency(table_FB)
print('dof=%d' % dof)
print('p=', p)
```

```python
# interpret test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
```

```python
# interpret p-value
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```
