## Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm
```

```python
X_u = FEV_upper_10[['age']]
Y_u = FEV_upper_10['FEV']
```

```python
# Chia dữ liệu thành 2 phần test và train
from sklearn.model_selection import train_test_split
```

```python
X_train_u, X_test_u, y_train_u, y_test_u = train_test_split(X_u, Y_u, test_size=0.2, random_state=1)
```

```python
# Train model
lm.fit(X_train_u, y_train_u)
```

```python
b = lm.intercept_
b
```

```python
m = lm.coef_[0]
m
```

#### y = mx + b

```python
# Predict trên X_train_u
yHat_train = lm.predict(X_train_u)
```

```python
# Predict trên X_test_u
yHat_test = lm.predict(X_test_u)
```

```python
reg_line = [(m * float(x)) + b for x in np.array(X_u)]
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_u, reg_line, color='orange', linewidth=3)
plt.scatter(X_train_u, y_train_u, color='red', label='Train Actual')
plt.scatter(X_train_u, yHat_train, color='green', label='Train Predict')
plt.scatter(X_test_u, y_test_u, color='black', label='Test Actual')
plt.scatter(X_test_u, yHat_test, color='blue', label='Test Predict')
plt.xlabel('Age')
plt.ylabel('FEV')
plt.legend(title='Notes:')
plt.show()
```

```python
# Find the MSE
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_u, lm.predict(X_u))
print('The mean square error of FEV and predicted value is: ', mse)
```

```python
# Find the R^2
print('The R-square is:', lm.score(X_u, Y_u))
```

```python
# Predict new data
X_new_u = pd.DataFrame(np.array([2, 3, 4, 5]))
y_new_u = lm.predict(X_new_u)
y_new_u
```

## Polinormial Regression

```python
def PlotPoly(model, independent_variable, dependent_variable, Name):
    x_new = np.linspace(1, 6, 360)
    y_new = model(x_new)
    
    plt.plot(independent_variable, dependent_variable, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Mathplotlib for FEV ~ Age')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898,))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('FEV')
    
    plt.show()
    plt.close()
```

```python
X_u_p = FEV_upper_10['age']
Y_u_p = FEV_upper_10['FEV']
```

```python
# X_u, Y_u
# Here we use a polynomial of the 3rd order
f = np.polyfit(X_u_p, Y_u_p, 3)
p = np.poly1d(f)
p
```

```python
f
```

```python
# Find R^2
from sklearn.metrics import r2_score
r_squared = r2_score(Y_u_p, p(X_u_p))
print('The R-square value of Polinomial Regression model is:', r_squared)
```

```python
# Find the MSE
mse = mean_squared_error(Y_u_p, p(X_u_p))
print('The mean square error of FEV and predicted value is: ', mse)
```

```python
plt.figure(figsize=(8,8))
PlotPoly(p, X_u_p, Y_u_p, 'age')
```

```python
# Predict new data
X_new_p = np.array([2, 3, 4, 5])
Y_new_p = p(X_new_p)
Y_new_p
```

```python
# Trực quan hóa dữ liệu
Y_u_p_pred = p(X_u_p)
```

```python
# Distribution plot
plt.figure(figsize=(8, 8))
ax1 = sns.distplot(Y_u, hist=False, color='r', label='Actual Values')
sns.distplot(Y_u_pred, hist=False, color='b', label='Fitted Values SR', ax=ax1)
sns.distplot(Y_u_p_pred, hist=False, color='g', label='Fitted Values PR', ax=ax1)

plt.title('Actual vs Fitted Values for FEV')
plt.xlabel('FEV')
plt.ylabel('freq')
plt.legend()
plt.show()
plt.close()
```

### PR tốt hơn SR => Có thể chọn PR
