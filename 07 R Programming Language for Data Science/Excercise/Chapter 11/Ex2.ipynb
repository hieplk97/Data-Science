```r
library("plotrix")
```

```r
Tree_species = c("Beech", "Chesnut", "Hawthorn", "Red maple", "Witch Hazel", "Yellow Birch")
number_of_tree = c(2, 2, 1, 4, 1, 1)
```

```r
pie_percent = round(100 * number_of_tree / sum(number_of_tree), 1)

pie3D(pie_percent, main = "Number of Trees of each Species",
      col = rainbow(length(pie_percent)))
```

```r
pie3D(pie_percent, labels = paste(pie_percent, "%"), main = "Number of Trees of each Species",
      col = rainbow(length(pie_percent)))

legend("topright", Tree_species, cex = 0.8, fill = rainbow(length(number_of_tree)))
```

```r
barplot(number_of_tree, names.arg = Tree_species, 
        xlab = "Species", ylab = "Number of trees", 
        col = "green", main = "Number of Trees of each Species", border = "red")
```
