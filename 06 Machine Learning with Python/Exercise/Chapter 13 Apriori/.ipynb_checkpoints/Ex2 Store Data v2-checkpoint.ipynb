### Cho dữ liệu store data trong tập tin dataset_group.csv.
### Yêu cầu: Áp dụng thuật toán Apriori để tính toán mức độ kết hợp giữa các item.
1. Chuẩn hóa dữ liệu
2. Áp dụng Apriori, Tìm kết quả
3. Tìm kiếm thông tin từ kết quả: trong thông tin kết quả có 'eggs' không? Nếu có thì 'eggs' kết hợp với item nào?"
4. Trực quan hóa dữ liệu
5. Cho biết 10 sản phẩm được mua nhiều nhất. Vẽ biểu đồ.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv("../../Data/dataset_group.csv", header = None, sep=',')
```

```python
data.info()
```

```python
data.head()
```

```python
df = data.iloc[:,1:3]
```

```python
df.head()
```

```python
dataset = df.groupby(1)[2].apply(list)
```

```python
dataset[1]
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
frequent_itemsets.head(3)
```

```python
frequent_itemsets.tail(3)
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.1)
rules
```

```python
# "Có eggs không? nó kết hợp với item nào?"
for row in rules.iterrows():
    if "eggs" in row[1][0]:
        print(row)
```

```python
support=rules['support'].values
confidence=rules['confidence'].values
lift = rules['lift'].values
```

```python
plt.scatter(support, confidence, s= lift*100,alpha=0.8,
            color=['red', 'green', 'blue', 'yellow', 'purple', 'orange'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
result = df.apply(pd.value_counts).fillna(0)
result
```

```python
df_true = result.iloc[1,:]
df_true[:10]
```

```python
x = df_true.sort_values(ascending=False)
```

```python
ten_products = x[:10]
ten_products
```

```python
pos = np.arange(len(ten_products.values))
```

```python
plt.bar(pos, ten_products.values, align='center')
plt.xticks(pos, ten_products.keys(), rotation='vertical')
plt.ylabel('Buying Times')
plt.xlabel('Names')
plt.title('Ten products best buy')
plt.show()
```
