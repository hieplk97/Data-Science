```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
f = open("Tai_nguyen_cung_cap/puma.txt", "r")
data = f.read()
```

```python
wc = WordCloud(background_color = 'white', max_words = 50)
wc.generate(data)

plt.figure(figsize = (10, 12))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image

wc_mask = np.array(Image.open('Tai_nguyen_cung_cap/puma_icon.jpg'))
wc1 = WordCloud(background_color = 'black', max_words = 50, mask = wc_mask)
wc1.generate(data)

plt.figure(figsize = (10, 12))
plt.imshow(wc1, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
