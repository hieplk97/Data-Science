```r
# Create data for the graph
x = c(46, 37, 35, 53)
labels = c("10A1", "10A2", "10A3", "10A4")
# Give the chart a file name
png(file = "classes_pie.jpg")
# Plot the chart
pie(x, labels)
# Save the file
dev.off()
```

```r
pie(x, labels, main = "Classes pie chart",
    col = rainbow(length(x)),
    clockwise = TRUE)
```

# Add percentage and legend

```r
pie_percent = round(100 * x / sum(x), 1)
# Plot the chart
pie(x, labels = paste(pie_percent, "%"), main = "Classes pie chart",
    col = rainbow(length(x)),
    clockwise = TRUE)
# insert legend
legend("topright", labels, cex = 0.8, fill = rainbow(length(x)))
```

# 3D pie chart

```r
library(plotrix)
```

```r
pie3D(x, labels = labels, main = "Classes Pie Chart",
      col = rainbow(length(x)))
```
