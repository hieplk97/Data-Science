```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# 1-1
heart_pd = pd.read_csv('tai_nguyen_cung_cap\heart.csv')
heart_pd.head()
```

```python
# 1-2
heart_pd.columns[heart_pd.isna().any()]
```

```python
# 1-3
heart_pd.drop(columns = ['ca', 'thal'])
```

```python
# 1-4
sb.boxplot(y = 'thalach', x = 'sex', data = heart_pd)
plt.show()
```

```python
# 1-6
heart_pd['chol'].fillna(heart_pd['chol'].interpolate(method='linear'), inplace = True)
```

```python
# 1-7
heart_pd['chol'].fillna(heart_pd['chol'].mean(), inplace=True)
```

```python
# 1-8
sb.scatterplot(y='chol', x='age', data=heart_pd)
plt.show()
```

```python
# 1-9
heart_pd.groupby(['slope', 'restecg'])['trestbps'].mean().reset_index()
```

```python
# 1-10
f, axes = plt.subplots(1, 2)
sb.regplot(x='age', y='thalach', data = heart_pd, ax=axes[0])
sb.regplot(x='age', y='trestbps', data = heart_pd, ax=axes[1], color='orange')
plt.show()
```

```python
# 1-11
age = heart_pd.groupby(['age']).size().reset_index(name='counts').sort_values(['counts'], ascending=False)
plt.title('So luot tuoi duoc khao sat nhieu nhat')
plt.xlabel('So tuoi')
plt.ylabel('So luot khao sat')
sb.barplot(x='age', y='counts', data=age.head(), color='blue')
plt.show()
```

```python
# 1-13
sex = heart_pd.groupby(['sex']).size().reset_index(name='counts')
plt.title('Ti le nam nu')
plt.pie(x = sex['counts'], labels=sex['sex'], autopct='%1.1f%%')
plt.show()
```
