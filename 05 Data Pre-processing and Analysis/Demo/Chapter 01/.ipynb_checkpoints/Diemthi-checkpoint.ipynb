```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import re
```

```python
diem_thi = pd.read_excel('Diemthi_thpt_quocgia_2016.xlsx')
diem_thi.info()
```

```python
diem_thi.head()
```

```python
danh_sach_mon_thi = ['Toán', 'Ngữ văn', 'Địa lí', 'Lịch sử', 'Tiếng Anh', 'Sinh học', 'Vật lí', 'Hóa học']

def tach_diem (diem):
    ds_diem = {}
    for mon_thi in danh_sach_mon_thi:
        search_str = mon_thi + ':\s\s\s\d+.\d+'
        kq = re.search(search_str, diem)
        if kq:
            temp = kq.group().partition(': ')
            ds_diem[mon_thi] = [float(temp[2])]
        else:
            ds_diem[mon_thi] = [np.nan]
    df_temp = pd.DataFrame(ds_diem)
    return df_temp
```

```python
for row_index in range(diem_thi.shape[0]):
    diem = diem_thi['DIEM_THI'][row_index]
    df_temp = tach_diem(diem)
    diem_thi = pd.concat([diem_thi, df_temp], ignore_index=True, sort=False)
```

```python
diem_thi.head()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.hist(diem_thi['Toán'].dropna())
plt.subplot(1, 2, 2)
plt.distplot(diem_thi['Toán'].dropna())
plt.show()
```
