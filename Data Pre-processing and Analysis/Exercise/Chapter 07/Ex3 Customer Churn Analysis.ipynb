```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
data.head()
```

```python
data.info()
```

```python
occ = data.Churn.value_counts()
occ
```

```python
# print the ratio of fraud cases
print(occ/len(data.index))
```

```python
X = data.drop(['customerID', 'Churn'], axis=1)
y = data.Churn

X.head()
```

```python
X.MonthlyCharges = X.MonthlyCharges.astype('float')
```

```python
X.TotalCharges = pd.to_numeric(X.TotalCharges, errors='coerce')
```

```python
y.head()
```

```python
# Chuẩn hóa y
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_new = label_encoder.fit_transform(y)
y_new[:5]
```

```python
# Chuẩn hóa X; các cột là category
# Categorical boolean mask
categorical_feature_mask = X.dtypes==object
# filter categorical columns using mask and turn it into a list
categorical_cols = X.columns[categorical_feature_mask].tolist()
categorical_cols
```

```python
X_new = pd.get_dummies(data=X, columns=categorical_cols, drop_first=True)
X_new.head()
```

```python
### Chia dữ liệu thành 2 bộ, tỷ lệ 80:20
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size=0.2)
```

```python
from collections import Counter
sorted(Counter(y_train).items())
```

### Vì dữ liệu mỗi lớp khá nhiều, tỷ lệ 3:1
### Có thể undersampling hoặc oversampling

```python
# Undersampling
from imblearn.under_sampling import RandomUnderSampler
```

```python
X_train = X_train.fillna(X_train.mean())
X_train.head()
```

```python
rus = RandomUnderSampler()
X_train_resampled, y_train_resampled = rus.fit_resample(X_train, y_train)
```

```python
sorted(Counter(y_train_resampled).items())
```

```python
# Oversampling
from imblearn.over_sampling import SMOTE
```

```python
X_train_S, y_train_S = SMOTE().fit_resample(X_train, y_train)
```

```python
sorted(Counter(y_train_S).items())
```
