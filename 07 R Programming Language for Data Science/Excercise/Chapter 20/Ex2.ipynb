```r
library(factoextra)
```

```r
data = read.csv("../../Data/shopping_data.csv")
```

```r
head(data)
```

```r
print(paste("cols:", ncol(data)))
print(paste("rows:", nrow(data)))
```

```r
# plot the chart
plot(x = data$Annual.Income..k.., y = data$Spending.Score..1.100., 
     xlab = "Annual Income", ylab = "Spending Score", main = "Annual Income vs Spending Score", pch = 15, col = "blue")
```

```r
wss = (nrow(data) -1) * sum(apply(data, 2, var))
for (i in 1:10) wss[i] = sum(kmeans(data[, 4:5], centers = i)$withinss)
```

```r
plot(1:10, wss, type = 'b', xlab = "Number of clustes", ylab = "WSS")
```

```r
set.seed(20)
dataCluster = kmeans(data[, 4:5], centers = 5, nstart = 20)
print(dataCluster)
```

```r
# plot the chart
dataCluster$cluster = as.factor(dataCluster$cluster)
plot(x = data$Annual.Income..k.., y = data$Spending.Score..1.100., 
     xlab = "Annual Income", ylab = "Spending Score", main = "Annual Income vs Spending Score", pch = 15,
     col = dataCluster$cluster)
lines(data$Annual.Income..k.., data$Spending.Score..1.100., col = 'orange', type = 'p')
```

```r
fviz_cluster(dataCluster, geom = "point",
             data = data[, 4:5]) + ggtitle("k = 5")
```
