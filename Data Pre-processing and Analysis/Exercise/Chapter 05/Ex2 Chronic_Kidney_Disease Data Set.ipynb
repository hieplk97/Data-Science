```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('data/chronic_kidney_disease.csv', header=None, names=['age', 'bp','sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba', 'bgr', 'bu', 'sc', 'sod', 
                                                                             'pot','hemo', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'class'])
print(dataset.shape)
dataset.info()
```

```python
dataset.head()
```

```python
dataset = dataset.replace('?', np.nan)
```

```python
dataset.columns
```

```python
# Chuyển chuỗi sang số
list_cols = ['age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc']
for col in list_cols:
    dataset[col] = pd.to_numeric(dataset[col])
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset  = dataset.dropna()
dataset.head()
```

```python
# Categorical boolean mask
categorical_feature_mask = dataset.dtypes==object
# filter categorical columns using mask and turn it into list
categorical_cols = dataset.columns[categorical_feature_mask].tolist()
categorical_cols
```
