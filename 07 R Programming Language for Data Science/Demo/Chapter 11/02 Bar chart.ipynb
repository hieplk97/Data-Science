```r
# Create the input vector
colors = c("green", "orange", "brown", "red")
classes = c("10A1", "10A2", "10A3")
grades = c("Excellent", "Very Good", "Good", "Medium")
```

```r
# Create the matrix of values
Values = matrix(c(5, 6, 7, 15, 14, 13, 20, 21, 22, 5, 4, 3),
                nrow = 4, ncol = 3, byrow = TRUE)
print(Values)
```

```r
# Create tthe bar chart
barplot(Values, main = "Grade of 3 classes", name.arg = classes,
        xlab = "Class", ylab = "Students", col = colors)
# Add legend to the chart
legend("right", grades, cex = 0.9, fill = colors)
```
