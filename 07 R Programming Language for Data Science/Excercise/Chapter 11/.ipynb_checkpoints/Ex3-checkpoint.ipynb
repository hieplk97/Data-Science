```r
year = c(2014, 2015, 2016, 2017)
first_fallen = c(309, 271, 263, 255)
one_hundred_fallen = c(309, 314, 312, 304)
```

```r
data = matrix(c(first_fallen, one_hundred_fallen), nrow = 2, ncol = 4, byrow = TRUE)

barplot(data, names.arg = year, 
        col=c("darkblue","red"), main = "First and last dates of leaf fall in Yellow Birch", beside = TRUE)
legend("bottomright", c("first fallen", "100% fall"), cex = 0.9, fill = c("darkblue","red"))
```
