```python
from sklearn.datasets import make_swiss_roll
```

```python
X, y = make_swiss_roll(n_samples=1000, noise=0.2, random_state=41)
```

```python
X
```

```python
y
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(16, 13))
ax = fig.add_subplot(2, 2, 1, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.hsv)
ax.view_init(10, -66)
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.set_ylabel("$x_3$", fontsize=18)
ax.set_title("Dữ liệu Swiss Roll", fontsize=22)
```

```python
from sklearn.manifold import LocallyLinearEmbedding
```

```python
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10, random_state=42)
X_reduced = lle.fit_transform(X)
```

```python
X_reduced
```

```python
plt.title("Dữ liệu Swiss Roll sau khi sử dụnh LLE", fontsize=22)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=plt.cm.hsv)
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$x_2$", fontsize=18)
plt.grid(True)
plt.show()
```
