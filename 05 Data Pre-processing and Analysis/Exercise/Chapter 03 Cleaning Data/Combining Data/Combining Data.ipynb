# Câu 5

```python
df5 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
                     'food': ['fish', 'beans', 'bread']}, columns=['name', 'food'])
df6 = pd.DataFrame({'name': ['Mary', 'Joseph'],
                     'drink': ['wine', 'beer']}, columns=['name', 'drink'])
display(df5, df6)
```

```python
df56_merge = pd.merge(df5, df6)
df56_merge
```

```python
df56_inner = pd.merge(df5, df6, how='inner')
df56_inner
```

```python
df56_outer = pd.merge(df5, df6, how='outer')
df56_outer
```

```python
df56_left = pd.merge(df5, df6, how='left')
df56_left
```

```python
df56_right = pd.merge(df5, df6, how='right')
df56_right
```

# Câu 6

```python
df7 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [1, 2, 3, 4]})
df8 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [3, 1, 2, 4]})
display(df7, df8)
```

```python
df78_on = pd.merge(df7, df8, on='name')
df78_on
```

```python
df78_on_suff = pd.merge(df7, df8, on='name', suffixes=['_L', '_R'])
df78_on_suff
```
