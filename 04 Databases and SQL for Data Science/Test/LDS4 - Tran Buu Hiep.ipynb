1-1. Liệt kê danh sách 10 customer của các country Poland, Argentina, USA, UK theo mẫu sau,
sắp tăng theo country, city (1 điểm)


```python
%sql SELECT c.Country, c.City, c.id, CONCAT(c.FirstName,' ',c.LastName) AS 'customer_name' FROM salesdb.customers c  WHERE c.Country IN ('USA', 'UK', 'Poland', 'Argentina') ORDER BY c.Country, c.City  LIMIT 10;
```

```python
%sql SELECT * FROM salesdb.customers WHERE customers.id NOT IN (SELECT DISTINCT orders.CustomerId FROM salesdb.orders) ORDER BY Country;
```

1-3. Liệt kê các đơn đặt hàng đặt trong tháng 3/2014 theo mẫu sau, sắp giảm dần theo
orderdate (0.5 điểm)

```python
%sql SELECT c.FirstName, o.OrderNumber, o.OrderDate, o.TotalAmount FROM salesdb.orders o JOIN salesdb.customers c ON o.CustomerId=c.Id WHERE YEAR(o.OrderDate) = '2014' AND MONTH(o.OrderDate) = '3' ORDER BY o.OrderDate DESC;
```

1-4. Liệt kê các đơn đặt hàng theo mẫu sau và có ngày đặt hàng từ ngày 7/9/2012 đến ngày
14/9/2012 (1 điểm)

```python
%sql SELECT o.OrderNumber, o.OrderDate, DAYNAME(o.OrderDate) AS 'day_of_week' FROM salesdb.orders o WHERE o.OrderDate BETWEEN '2012/09/07' AND '2012/09/14';
```

1-5. Liệt kê các đơn đặt hàng theo mẫu sau và có ngày đặt hàng là ngày chủ nhật của tháng
3/2014, sắp giảm theo orderdate (1 điểm)

```python
%sql SELECT o.OrderNumber, o.OrderDate, CONCAT(c.FirstName,' ',c.LastName) AS 'customer_name', o.TotalAmount FROM salesdb.orders o JOIN salesdb.customers c ON o.CustomerId=c.Id WHERE YEAR(o.OrderDate)='2014' AND MONTH(o.OrderDate)='3' AND DAYNAME(o.OrderDate)='Sunday' ORDER BY o.OrderDate DESC;
```

1-6. Liệt kê các customer theo mẫu sau và đếm các đơn đặt hàng của các customer, chỉ in các
customer nào có đếm>15, sắp giảm theo đếm (1 điểm) 

```python
%sql SELECT CONCAT(c.FirstName,' ',c.LastName) AS 'customer_name', c.Country, c.City, COUNT(*) AS 'count_order' FROM salesdb.customers c JOIN salesdb.orders o ON c.Id=o.CustomerId GROUP BY c.Id  HAVING COUNT(*) > 15 ORDER BY `count_order` DESC;
```

1-7. Thống kê các product theo mẫu sau, gồm các thông tin : id, productname, sum quantity,
avg unitprice. Chỉ in các product có sum quantity từ 800 đến 1000, sắp giảm theo avg
unitprice (1 điểm) 

```python
%sql SELECT p.Id, p.ProductName, SUM(i.Quantity) AS 'Sum_quantity', AVG(i.UnitPrice) AS 'avg_price' FROM salesdb.products p JOIN salesdb.orderitems i ON p.Id=i.ProductId GROUP BY p.Id HAVING SUM(i.Quantity) BETWEEN '800' AND '1000' ORDER BY `avg_price` DESC;
```

1-8. Cho biết product nào được đặt hàng nhiều nhất. In theo mẫu sau (0.5 điểm)

```python
%sql SELECT p.Id, p.ProductName, COUNT(i.Id) AS 'count_product' FROM salesdb.products p JOIN salesdb.orderitems i ON p.Id=i.ProductId GROUP BY p.Id ORDER BY `count_product` DESC LIMIT 1;
```

1-9. Liệt kê theo mẫu sau gồm 10 đơn đặt hàng có đặt product tên là Chai, sắp giảm theo
totalamount (1 điểm) 

```python
%sql SELECT o.* FROM salesdb.orders o JOIN salesdb.orderitems i ON o.Id=i.OrderId JOIN salesdb.products p ON i.ProductId=p.Id WHERE p.ProductName='Chai' ORDER BY o.TotalAmount DESC LIMIT 10;
```

1-10. Liệt kê theo mẫu sau gồm các đơn đặt hàng có totalamount thấp nhất (min) và cao nhất
(max) (0.5 điểm) 

```python
%sql SELECT o.OrderNumber, o.OrderDate, o.TotalAmount, 'min total amount' AS 'Note' FROM salesdb.orders o WHERE o.TotalAmount = (SELECT MIN(o.TotalAmount) FROM salesdb.orders o) UNION SELECT o.OrderNumber, o.OrderDate, o.TotalAmount, 'max total amount' AS 'Note' FROM salesdb.orders o WHERE o.TotalAmount = (SELECT MAX(o.TotalAmount) FROM salesdb.orders o);
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

2-1. Vẽ biểu đồ bar như sau, gồm 10 customer có nhiều đơn đặt hàng nhất (1 điểm)

```python
data = %sql SELECT c.FirstName, COUNT(*) AS 'count_order' FROM salesdb.customers c JOIN salesdb.orders o ON c.Id=o.CustomerId GROUP BY c.Id ORDER BY `count_order` DESC LIMIT 10;
df1 = data.DataFrame()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(data=df1, x='FirstName', y='count_order', ci=None, color='green')
plt.xticks(rotation=45)
plt.ylabel('Số đơn đặt hàng')
plt.xlabel('Tên khách hàng')
plt.show()
```

2-2. Vẽ 2 biểu đồ bar như sau. Biểu đồ 1 vẽ theo các năm và đếm số đơn đặt hàng của mỗi năm.
Biểu đồ 2 vẽ theo các năm và tính tổng totalamount của mỗi năm (1 điểm)

```python
data = %sql SELECT YEAR(o.orderDate) as 'year', COUNT(*) AS 'order_count', SUM(o.TotalAmount) AS 'total' FROM salesdb.orders o GROUP BY YEAR(o.orderDate);
df2 = data.DataFrame()
```

```python
plt.figure(figsize=(14, 5))


plt.subplot(1, 2, 1)
sns.barplot(data=df2, x='year', y='order_count', ci=None, color='blue')
plt.ylabel('Số đơn ĐH')
plt.xlabel('Năm')
plt.title('1st Chart')

plt.subplot(1, 2, 2)
sns.barplot(data=df2, x='year', y='total', ci=None, color='blue')
plt.ylabel('Tổng giá trị')
plt.xlabel('Năm')
plt.title('2nd Chart')
plt.grid()

plt.show()
```
