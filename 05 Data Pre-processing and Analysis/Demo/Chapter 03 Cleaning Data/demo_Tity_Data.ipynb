```python
import pandas as pd
```

```python
df = pd.read_csv('tidy_data.csv')
df
```

```python
# Biến đổi sang dạng Tidy Data bằng cách dử dụng pd.melt
df_tidy1 = pd.melt(frame=df, id_vars='Name', value_vars=['Treatment a', 'Treatment b'])
df_tidy1
```

```python
#đặt tên mới cho cột biến đổi
df_tidy2 = pd.melt(frame=df, id_vars='Name', value_vars=['Treatment a', 'Treatment b'], 
                   var_name='Treatment', value_name= 'Result')
df_tidy2
```

```python
# Biến đổi từ Tidy Data sang dạng Report Data
df_pivot = df_tidy2.pivot(index='Name', columns='Treatment', values='Result')
df_pivot
```

```python
df_pivot.reset_index(inplace=True)
df_pivot
```

```python
df_tidy3 = pd.melt(frame=df, id_vars='Name', value_vars=['Treatment a', 'Treatment b'], 
                   var_name='Treatment', value_name= 'Result')
df_tidy3
```

```python
df_tidy3 = df_tidy3.append({'Name':'Jane', 'Treatment': 'Treatment b', 'Result': 30}, ignore_index=True)
df_tidy3
```

```python
df_pivot2 = df_tidy3.pivot(index='Name', columns='Treatment', values='Result')
df_pivot2
```
