```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tb-raw.csv')
df.head()
```

## Vấn đề:
* Tên cột chứa giá trị, chứa thông tin giới tính và độ tuổi
* Trong cell chứa giá trị hỗn hợp (value, 0, Nan)

```python
# Chuyển dữ liệu theo định dạng country, year và sex_and_age
df_after = pd.melt(df, id_vars=['country', 'year'], value_name='cases', var_name='sex_and_age')
df_after.head()
```

```python
# Tách cột sex_and_age thành age và sex
tmp_df = df_after['sex_and_age'].str.extract('(\D)(\d+)(\d{2})')

# Tên các cột
tmp_df.columns = ['sex', 'age_lower', 'age_upper']

# Tạo cột age
tmp_df['age'] = tmp_df['age_lower'] + '-' + tmp_df['age_upper']
```
