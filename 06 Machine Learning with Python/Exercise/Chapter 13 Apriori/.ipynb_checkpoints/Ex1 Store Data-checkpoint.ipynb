### Cho dữ liệu store data trong tập tin store_data.
### Yêu cầu: Áp dụng thuật toán Apriori để tính toán mức độ kết hợp giữa các item.
1. Chuẩn hóa dữ liệu
2. Áp dụng Apriori, Tìm kết quả
3. Tìm kiếm thông tin từ kết quả: trong thông tin kết quả có 'milk' không? Nếu có thì 'milk' kết hợp với item nào?"
4. Trực quan hóa dữ liệu
5. Cho biết 10 sản phẩm được mua nhiều nhất, vẽ biểu đồ biểu diễn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# load dữ liệu
store_data = pd.read_csv('../../Data/store_data.csv', header= None)
```

```python
store_data.info()
```

```python
store_data.head(3)
```

```python
records = []
for i in range(0, store_data.shape[0]):
    records.append([str(store_data.values[i,j]) for j in range(0, store_data.shape[1])])
```

```python
records[0]
```

```python
te = TransactionEncoder()
te_ary = te.fit(records).transform(records)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.shape
```

```python
df.info()
```

```python
df.head(3)
```

```python
df.tail()
```

```python
df.columns
```

```python
df = df.drop(['nan'], axis=1)
```

```python
frequent_itemsets = apriori(df, min_support=0.03, use_colnames=True)
frequent_itemsets.head()
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.4)
rules
```

```python
print(rules.info())
```

```python
# "Có milk không? nó kết hợp với item nào?"
for row in rules.iterrows():
    if "milk" in row[1][0]:
        print(row)
```

```python
support=rules['support'].values
confidence=rules['confidence'].values
lift = rules['lift'].values
```

```python
plt.figure(figsize=(8,5))
plt.scatter(support, confidence, s= lift*100,alpha=0.8, c = lift)
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
result = df.apply(pd.value_counts).fillna(0)
result
```

```python
df_true = result.iloc[1,:]
df_true[:10]
```

```python
x = df_true.sort_values(ascending=False)
```

```python
ten_products = x[:10]
ten_products
```

```python
pos = np.arange(len(ten_products.values))
```

```python
plt.figure(figsize=(8,4))
plt.bar(pos, ten_products.values, align='center')
plt.xticks(pos, ten_products.keys(), rotation='vertical')
plt.ylabel('Buying Times')
plt.xlabel('Names')
plt.title('Ten products best buy')
```
