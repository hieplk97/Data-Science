```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame({
    'x': np.random.randint(-100, 100, 1000).astype(float),
    'y': np.random.randint(-80, 80, 1000).astype(float),
    'z': np.random.randint(-150, 150, 1000).astype(float)
})

df.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
f = plt.figure(figsize=(8,8))
ax = f.add_subplot(111, projection='3d')
ax.scatter(df['x'], df['y'], df['z'])
plt.show()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
scaler = Normalizer()

df_scale = scaler.fit_transform(df)
df_scale = pd.DataFrame(df_scale, columns=['x', 'y', 'z'])
```

```python
f = plt.figure(figsize=(16,8))
ax1 = f.add_subplot(121, projection='3d')
ax1.scatter(df['x'], df['y'], df['z'])
ax2 = f.add_subplot(122, projection='3d')
ax2.scatter(df_scale['x'], df_scale['y'], df_scale['z'])
plt.show()
```
