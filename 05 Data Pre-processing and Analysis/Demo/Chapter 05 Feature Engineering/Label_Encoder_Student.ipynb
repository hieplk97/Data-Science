```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df_sv = pd.DataFrame({'Name': ['John', 'Lucy', 'Lily', 'Mark', 'Tom'],
                      'Sex': ['Male', 'Female', 'Other', 'Male', 'Male'],
                       'Age': [16, 17, 16, 17, 15],})
df_sv
```

```python
df_sv['Stage'] = ['Second', 'First', 'Third', 'Second', 'Four']
df_sv
```

```python
encoder = LabelEncoder()
df_sv['Stage_New'] = encoder.fit_transform(df_sv.Stage)
df_sv
```
