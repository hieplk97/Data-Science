```python
# Câu 1: Vẽ scatterplot: Phân loại với các điểm không chồng chéo, so sánh Attack (trục x) & Defense (trục y)
# sort dữ liệu theo 'Attack', lấy 100 dòng đầu
plt.figure(figsize = (10, 8))
sb.swarmplot(x = 'Attack', y = 'Defense', data = df[:100])
plt.show()
```

```python
# Câu 7: Vẽ barplot: Nhóm dữ liệu theo Type 1, cho biết mỗi type có mấy mẫu, sắp tăng dần
result = df.groupby('Type 1').HP.count()
type(result)
```

```python
df_result = pd.DataFrame(result)
df_result['label'] = df_result.index
df_result = df_result.sort_values(by = ['HP'])
df_result.info()
```

```python
plt.figure(figsize = (8, 6))
x = sb.barplot(x = 'label', y = 'HP', data = df_result)
x.set_xticklabels(df_result.label, rotation = 45)
plt.show()
```

```python
# Câu 8: Vẽ pairplot: thể hiện mối quan hệ giữa "Attack", "Defense" theo từng 'Type 1', sử dụng lại dữ liệu của câu 5
sb.pairplot(data = stats_df,
            vars = ["Attack", "Defense"],
            kind = 'reg', 
            palette = 'bright',
            diag_kind = 'kde',
            hue = 'Type 1',
            height = 4,
            aspect = 2)
plt.show()
plt.clf()
```

```python
# Câu 9: Vẽ JointGrid: gồm sb.regplot, sb.distplot để thể hiện mối quan hệ giữa "Attack", "Defense"
g = sb.JointGrid(data = df, x = 'Attack', y = 'Defense', height = 8)
g.plot(sb.regplot, sb.distplot)
plt.show()
plt.clf()
```
