```r
kings = scan("../../Data/Kings.dat.txt", skip = 3)
print(kings)
```

```r
# Convert tot time series object
kings.timeserries = ts(kings, start = 1, end = length(kings), frequency =1)
```

```r
print(kings.timeserries)
```

```r
plot(kings.timeserries, main = "Age of Death of Successive Kings of England", col = 'blue', xlab = 'Kings', ylab = 'Age')
```
