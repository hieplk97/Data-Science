```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(1)
df = pd.DataFrame({
    'X1': np.random.normal(loc=0, scale=2, size=10000),
    'X2': np.random.normal(loc=5, scale=3, size=10000),
    'X3': np.random.normal(loc=-5, scale=5, size=10000)
})

df.head()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1,3,1)
sns.distplot(df['X1'])

plt.subplot(1,3,2)
sns.distplot(df['X2'])

plt.subplot(1,3,3)
sns.distplot(df['X3'])

plt.show()
```

```python
df['X1'].mean(), df['X1'].median(), df['X1'].skew()
```

```python
f, ax = plt.subplots(ncols=1, figsize=(8, 6))
sns.distplot(df['X1'], ax=ax)
sns.distplot(df['X2'], ax=ax)
sns.distplot(df['X3'], ax=ax)

plt.show()
```

```python
f, ax = plt.subplots(ncols=1, figsize=(8, 6))
sns.kdeplot(df['X1'], ax=ax)
sns.kdeplot(df['X2'], ax=ax)
sns.kdeplot(df['X3'], ax=ax)

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_scale = scaler.fit_transform(df)
df_scale = pd.DataFrame(df_scale, columns=['X1', 'X2', 'X3'])
df_scale.head()
```

```python
f, ax = plt.subplots(ncols=1, figsize=(8, 6))
sns.kdeplot(df_scale['X1'], ax=ax)
sns.kdeplot(df_scale['X2'], ax=ax)
sns.kdeplot(df_scale['X3'], ax=ax)

plt.show()
```
