```r
library("forecast")
```

```r
rainfall = scan("../../Data/rain_fall.txt")
```

```r
ts.rainfall = ts(rainfall, frequency = 12, start = c(2014, 1))
```

```r
forecast.rainfall = HoltWinters(x = ts.rainfall, beta = F, gamma = F)
```

```r
print(paste("Sum Square Error", forecast.rainfall$SSE))
```

```r
plot(forecast.rainfall)
```

# Simple Exponetial Smoothing (SES)

```r
forecast.HoltWinters_new = forecast(forecast.rainfall, h = 5)
```

```r
forecast.HoltWinters_new
```

```r
plot(forecast(forecast.HoltWinters_new))
```

# Holt Exponential Smoothing (HES)

```r
HES.forecast = HoltWinters(log(ts.rainfall), gamma = F)
```

```r
HES.forecast
```

```r
print(paste("Sum Square Error", HES.forecast$SSE))
```

```r
plot(HES.forecast)
```

```r
HES.HoltWinters_new = forecast(HES.forecast, h = 5)
```

```r
plot(forecast(HES.HoltWinters_new))
```

### => HES Khong thich hop

```r
y = data.frame(HES.HoltWinters_new)
```

```r
y$Forecast = exp(y$Point.Forecast)
```

```r
y
```

# Holt Winters

```r
# Use log to optimization
HWES.forecast = HoltWinters(log(ts.rainfall))
```

```r
HWES.forecast
```

```r
plot(HWES.forecast)
```

```r
HWES.forecast_new = forecast(HWES.forecast, h = 5)
```

```r
HWES.forecast_new
```

```r
print(paste("Sum Square Error", HWES.forecast$SSE))
```

```r
x = data.frame(HWES.forecast_new)
x
```

```r
x$Forecast = exp(x$Point.Forecast)
x
```
