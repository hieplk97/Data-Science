```python
import pandas as pd
```

```python
df = pd.read_csv('../data/weather-raw.csv')
df.head()
```

## Vấn đề:
* Biến được lưu trữ trong cả các dòng (min, max) và các cột (days)
* Tên cột chứa giá trị day thay vì biến
* Trong từng cell chứa tempurature nhưng có rất nhiều giá trị NaN

=> Di chuyển 3 biến bị đặt sai (tmin, tmax và day) thành 3 cột riêng lẻ

```python
df_after = pd.melt(df, id_vars=['id', 'year', 'month', 'element'], var_name='day_raw')
df_after.head()
```

```python
# Tạo cột day chỉ chứa ngày
df_after['day'] = df_after['day_raw'].str.extract('d(\d+)', expand=False)
df_after.head()
```

```python
# Chuyển dữ liệu thành số
df_after[['year', 'month', 'day']] = df_after[['year', 'month', 'day']].apply(lambda x: x.astype(int))
df_after.head()
```

```python
# Tạo cột date chứa cả day, month, year
import datetime


def create_date(row):
    return datetime.datetime(year=row['year'], month=row['month'], day=row['day'])
```

```python
df_after['date'] = df_after.apply(lambda row: create_date(row), axis=1)
```

```python
# Bỏ các cột thừa
df_after = df_after.drop(['year', 'month', 'day', 'day_raw'], axis=1)
```

```python
# Bỏ các dòng dữ liệu NaN
df_after =df_after.dropna()
df_after.head()
```
