## Chapter7 - Exercise 1: Vẽ Barplot, Swarmplot

#### Thực hiện các yêu cầu sau, và đối chiếu với kết quả cho trước:

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
# Câu 1: Cho dữ liệu:
x =  ["Excellent","Very Good","Good","Medium"]
y = [5.0,15.0,20.0,5.0]

data = dict(zip(x, y))
# Hãy vẽ barplot
sb.barplot(x,y)
plt.xlabel('Xếp loại')
plt.ylabel('Tổng số')
plt.show()
```

```python
# Câu 2: Cho dữ liệu
chars = "A A A A A A A A A A B B B B B B C C C C C C D D D D D D D D D D E E E E E E E F F F F G G G G G G G H H H H"
# Hãy vẽ barplot với mỗi cột có tên là ký tự, giá trị là số lần xuất hiện của ký tự đó trong chuỗi
x = list(set(chars))
x.remove(' ')
x.sort()
y = []

for word in x:
    y.append(chars.count(word))
    
sb.barplot(x,y)
plt.xlabel('Giá trị')
plt.ylabel('Ký tự')
plt.show()
```

```python
# Câu 3: Biểu diễn dữ liệu data trên bằng swarmplot
sb.swarmplot(x,y)
plt.xlabel('Giá trị')
plt.ylabel('Ký tự')
plt.show()
```
