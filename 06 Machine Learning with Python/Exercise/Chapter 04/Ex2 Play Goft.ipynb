```python
import pandas as pd
import numpy as np
# Xuất model
import pickle

from sklearn.preprocessing import LabelEncoder
from sklearn.utils.validation import column_or_1d
# Sử dụng BernoulliNB
from sklearn.naive_bayes import BernoulliNB
```

```python
df = pd.read_excel('../../Data/playgolf_data.xlsx',engine='openpyxl', index_col = 0)
df.head()
```

```python
features = df.drop("Play Golf", axis=1)
target = df[["Play Golf"]]
```

```python
features = pd.get_dummies(features)
features
```

```python
le = LabelEncoder()
target = le.fit_transform(column_or_1d(target))
```

```python
model1 = BernoulliNB()
model1.fit(features, target)
```

```python
# Kiểm tra độ chính xác
print("The prediction accuracy is: ",
model1.score(features,target)*100,"%")
```

```python
class_names = model1.classes_
class_names
```

```python
# Save to file in the current working directory
pkl_filename = "playgoft_model.pkl"
with open(pkl_filename, 'wb') as file:
    pickle.dump(model1, file)
```

```python
with open(pkl_filename, 'rb') as file:
    playgoft_model = pickle.load(file)
```
